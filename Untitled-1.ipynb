```python
import pandas as pd
```

```python
df = pd.read_csv("forward_filled_dataset_1.csv")        
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.index = df['date']
del df['date']
```

```python
df['next_day_mean_temp'] = df.shift(-1)['mean_temp']
df = df.iloc[:-1, :].copy()
```

```python

df['wind_speed'] = df['wind_speed'].ffill()
df.isna().sum()
```

```python
df
```

```python
df.to_csv('with_target.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df
```

```python
X = df[['cloud_cover','wind_speed','wind_gust','humidity','pressure','radiation','precipitation','snow_depth','sunshine', 'mean_temp']]
y = df['next_day_mean_temp']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_predicted)
r2 = r2_score(y_test, y_predicted)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")
```

```python
model.coef_
#model.intercept_
```

```python
y_predicted = pd.DataFrame(y_predicted)
y_predicted.index = y_test.index
```

```python
results = pd.concat([y_test, y_predicted], axis=1)
results.columns = ["actual", "predicted"]
results
```

```python
results.plot()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
from sklearn import tree
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("with_target.csv")
```

```python
X = df[['cloud_cover','wind_speed','wind_gust','humidity','pressure','radiation','precipitation','snow_depth','sunshine']]
y = df['next_day_mean_temp']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model = tree.DecisionTreeRegressor()
model.fit(X_train, y_train)
y_predicted = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_predicted)
r2 = r2_score(y_test, y_predicted)

print(f"Mean Squared Error: {mse}")
print(f"R² Score: {r2}")
```

```python
y_predicted = pd.Series(y_predicted, index=y_test.index)
rr = pd.concat([y_test, y_predicted], axis=1)
```

```python
rr
```

```python
rr.columns = ["actual", "calculated"]
rr.plot(figsize=(16, 9))
```
